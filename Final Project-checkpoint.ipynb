## Problem 1: Correlation between Average Exit Velo and Average Home Run Distance

```python
#Correlation Matrix to see columns to make scatter plots out 
ev_la_corr_matrix = baseball_data.corr(numeric_only = True)
plt.matshow(ev_la_corr_matrix)
plt.colorbar()
plt.xticks(range(20), baseball_data.columns, rotation = 90)
plt.yticks(range(20), baseball_data.columns, rotation = 0)
```

```python
#Create the First graph without a line of best fit
baseball_data.plot.scatter("average_ev", "average_homerun", marker = "^", s = 20, color = "purple")
#Define the x and y labels, and the title 
plt.xlabel("Average Exit Velocity", fontsize = 14)
plt.ylabel("Average Homerun Distance", fontsize = 14)
plt.title("Average Homerun Distance vs Average Exit Velocity", fontsize = 18)
```

Based upon the given graph there shows to be a bunch of data in a tight cluster ranging from 85-93 exit velo and 380-420 feet.
Majority of the data sits within those incraments.

```python
#Define average_ev and average_homerun
average_ev = baseball_data['average_ev']
average_homerun = baseball_data['average_homerun']

#Find the Correlation between ev and homerun, print the correlation
ev_corr = scipy.stats.pearsonr(average_ev, average_homerun)
print("Correlation between Average Exit Velocity and Average Homerun Distance: ", ev_corr[0])
```

```python
#Create the Linear Regrssion model
ev_linear_model = scipy.stats.linregress(average_ev, average_homerun)
print(ev_linear_model)
```

```python
#Define and print the slope
ev_slope = ev_linear_model.slope
print("Slope: ", ev_slope)
#Define and print the y intercept 
ev_y_int = ev_linear_model.intercept
print("Y-intercept: ", ev_y_int)

#Make the equation for the linear fit line
ev_linear_fit = ev_slope * average_ev + ev_y_int
```

```python
#Graph the model with the line of best fit and correlation score
plt.scatter(average_ev, average_homerun, s = 20, marker = "^", color = "purple")
plt.plot(average_ev, ev_linear_fit, linewidth = 3, linestyle = ":", label = "Best Fit", color = "green")
#Make the x and y labels, along with a title
plt.xlabel("Average Exit Velocity (mph)", fontsize = 14)
plt.ylabel("Average Homerun Distance (ft)", fontsize = 14)
plt.title("Avg Homerun Distance vs Avg Exit Velocity", fontsize = 20)
#Display the legend
plt.legend(loc = "upper left")
```

The graph here now displays the influence that Average Exit Velocity (mph) has on Average Homerun Distance (feet), with a line of best fit included.
The slope of the line shows an upward trend, meaning for the majority of the data, the harder the ball is hit the farther the ball goes. 
The majority of the data still sits in between the same ranges as before.

```python
#Crosstab average exit velocity and home run distance for chi sqaured test
average_ev_homerun = pd.crosstab(baseball_data['average_ev'], baseball_data['average_homerun'], normalize = 'columns')
```

```python
c, p, dof, expected = scipy.stats.chi2_contingency(average_ev_homerun)
print("The P-value is: ", p)
```

```python
print(average_ev_homerun - expected)
```

## Problem 3: Unit of Power

```python
def power(average_ev, hard_hit_percentage, batted_ball_events):
    power = ((average_ev * hard_hit_percentage) / batted_ball_events) * 1.73448503    
    return power
```

```python
print("2022 --> Aaron Judge Power: ", power(95.8, 61.6, 341))
```

```python
print("2015 --> Billy Hamilton Power: ", power(83.7, 8.9, 350))
```

```python
print("2018 Randal Grichuk Power: ", power(90.7, 43.4, 305))
```
